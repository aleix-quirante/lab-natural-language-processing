## Now, we have to clean the html code removing words

- First we remove inline JavaScript/CSS
- Then we remove html comments. This has to be done before removing regular tags since comments can contain '>' characters
- Next we can remove the remaining tags

```python
import re

def remove_html(text):
    return re.sub(r'<[^>]+>', "", text)

```

- Remove all the special characters
    
- Remove numbers
    
- Remove all single characters
 
- Remove single characters from the start

- Substitute multiple spaces with single space

- Remove prefixed 'b'

- Convert to Lowercase

```python
def clean_text(text):
    # 1. HTML
    text = re.sub(r'<[^>]+>', "", text)
    # 2. Lowercase
    text = text.lower()
    # 3. Numbers
    text = re.sub(r'\d+', '', text)
    # 4. Special Characters
    text = re.sub(r'[^\w\s]', ' ', text)
    # 5. Single chars (middle)
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    # 6. Single chars (start)
    text = re.sub(r'^[a-zA-Z]\s+', ' ', text)
    # 7. Prefixed 'b'
    text = re.sub(r'^b\s+', '', text)
    # 8. Multiple spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text
```

## Now let's work on removing stopwords
Remove the stopwords.

```python
# Load the list of english stop words 
stop_words_list = set(stopwords.words('english'))

def remove_stopwords(text):
    # 1. Split the text into individual words
    # 2. Keep only the words that are NOT in our stop words list
    clean_words = [w for w in text.split() if w not in stop_words_list]
    # 3. Join them back into a sentence
    return " ".join(clean_words)

# Apply it to the dataframe
data["text"] = data["text"].apply(remove_stopwords)

# Check the result
print(data.head())
```
